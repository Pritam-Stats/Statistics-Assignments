### **2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)**

```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

data = np.random.randint(1, 7, size=1000)
counter = Counter(data)

x = list(counter.keys())    #data points
y = [counter[k]/1000 for k in x]    #prob

plt.bar(x, y)
plt.title('Discrete Uniform PMF (Unif(1,6))')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

### **3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution**

```python
def bernoulli_pdf(p, x):
    if x not in [0, 1]:
        return 0
    return p if x == 1 else (1 - p)

# Example
print('P(X=1):', bernoulli_pdf(0.7, 1))
print('P(X=0):', bernoulli_pdf(0.7, 0))
```

### **4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram**

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

n, p = 10, 0.5
data = binom.rvs(n, p, size=1000)

plt.hist(data, bins=np.arange(12)-0.5, density=True, edgecolor='black')
plt.title('Binomial Distribution Histogram (n=10, p=0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### **6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution**

### **9. Write a Python function to calculate Z-scores from a dataset and plot them**

```python
def z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

data = np.random.normal(loc=10, scale=2, size=100)
z = z_scores(data)

plt.hist(z, bins=20, edgecolor='black')
plt.title('Z-scores Distribution')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### **10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution**

```python
# Central Limit Theorem demo with exponential distribution

sample_means = []
for _ in range(1000):
    sample = np.random.exponential(scale=2, size=30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, density=True, edgecolor='black')
plt.title('CLT Demonstration using Exponential Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

### **11. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem**

```python
import numpy as np
import matplotlib.pyplot as plt

sample_means = []
for _ in range(1000):
    sample = np.random.normal(loc=10, scale=2, size=30)
    sample_means.append(np.mean(sample))

plt.hist(sample_means, bins=30, density=True, edgecolor='black')
plt.title('CLT with Normal Distribution Samples')
plt.xlabel('Sample Means')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

### **15. Implement hypothesis testing using Z-statistics for a sample dataset**

```python
from scipy.stats import norm

data = np.random.normal(loc=50, scale=10, size=30)
mu0 = 52
z_stat = (np.mean(data) - mu0) / (np.std(data, ddof=1) / np.sqrt(len(data)))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f'Z-statistic: {z_stat:.4f}, P-value: {p_value:.4f}')
```

### **16. Create a confidence interval for a dataset using Python and interpret the result**

```python
import scipy.stats as stats

data = np.random.normal(60, 5, 100)
mean = np.mean(data)
se = stats.sem(data)
ci = stats.norm.interval(0.95, loc=mean, scale=se)
print(f'95% Confidence Interval: {ci}')
```

### **17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean**

### **21. Write a Python program to compare the standard deviations of two datasets and visualize the difference**

```python
data1 = np.random.normal(50, 5, 100)
data2 = np.random.normal(50, 15, 100)

print(f'STD of data1: {np.std(data1):.2f}')
print(f'STD of data2: {np.std(data2):.2f}')

plt.hist(data1, alpha=0.5, label='STD=5')
plt.hist(data2, alpha=0.5, label='STD=15')
plt.legend()
plt.title('Comparison of Standard Deviations')
plt.grid(True)
plt.show()
```

### **22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution**

```python
data = np.random.normal(70, 10, 1000)
range_val = np.max(data) - np.min(data)
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1
print(f'Range: {range_val:.2f}, IQR: {iqr:.2f}')
```

### **23. Implement Z-score normalization on a dataset and visualize its transformation**

```python
data = np.random.normal(80, 20, 100)
z = (data - np.mean(data)) / np.std(data)

plt.hist(z, bins=20, edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.grid(True)
plt.show()
```

### **24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution**

```python
from scipy.stats import skew, kurtosis

data = np.random.normal(0, 1, 1000)
skew_val = skew(data)
kurt_val = kurtosis(data)

print(f'Skewness: {skew_val:.4f}, Kurtosis: {kurt_val:.4f}')
```
